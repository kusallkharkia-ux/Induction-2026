```python
import cv2
import numpy as np
img = cv2.imread('/Users/kushalkharkia/anaconda_projects/Preprocessing/WhatsApp Image 2026-04-08 at 21.25.12.jpeg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv, np.array([0, 100, 40]), np.array([10, 255, 255]))
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
if contours:
    c = max(contours, key=cv2.contourArea)
    maxy = min(c[:, :, 1])[0]
    top = c[c[:, :, 1] < (maxy + 25)]
    cv2.polylines(img, [top], False, (0, 255, 0), 3)
    M = cv2.moments(top)
    if M["m00"] != 0:
        cX = int(M["m10"] / M["m00"])
        cY = int(M["m01"] / M["m00"])
        cv2.circle(img, (cX, cY), 7, (0, 0, 255), -1)
        print(f"Rim Center: X={cX}, Y={cY}")
cv2.imshow('Rim Border', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
